## If a column has missing value and datatype int it will be automatically converted into float such as am column below and so on

```python
df=pd.read_csv('mtcarsmissing.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().head()
df.isnull().sum()# counting missing values in each column
```

# Filna

```python
# columns that has missing values we can give them values as every missing value
df.fillna({
    'cyl':4,
    'disp':120,
    'hp':150,
    'drat':3.9,
    'wt':2.0,
    'qsec':16.0,
    'vs':0,
    'am':1,
    'gear':4,
})
```

```python
df.head()
```

# Forwardfill
## value that is place above row will assign to the missing entry

```python
df.fillna(method='ffill').head()
```

```python
df.head()
```

# Backwardfill
## value that is place below row will assign to the missing entry

```python
df.fillna(method='bfill').head()
```

# Replace missing data with average value

```python
df.fillna(df.mean()).head()
```

```python
df.head()
```

```python
df.dropna().head()
```
