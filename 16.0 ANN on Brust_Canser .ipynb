```python
import pandas as pd
```

```python
df=pd.read_csv('breast_cancer.csv')
df.head()
```

```python
X=df.iloc[:,:-1]
Y=df.outcome
```

```python
X.describe()
```

## set range for the data to avoid overfitting
## feature_range=(-1,1) min is -1 and max is 1

```python
from sklearn.preprocessing import MinMaxScaler#MinMaxScaler convert data to similar range 
scaler=MinMaxScaler(feature_range=(-1,1))
X_scale=scaler.fit_transform(X)# fit_transform build autmatically array
```

```python
X_scale=pd.DataFrame(X_scale,columns=X.columns)# convert current array to pandas
```

```python
X_scale.head()
```

```python
X_scale.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_scale,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.neural_network import MLPClassifier# multi layer poceptron
model=MLPClassifier(hidden_layer_sizes=(128,128,64),max_iter=300,shuffle=False)# Hiden Layer Neurons (hidden_layer_sizes=(64,64,32) 3 hiden layers
#shuffle=False work as random state
model.fit(X_train,Y_train)
```

```python
Y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test, Y_predict)
```

```python
pd.crosstab(Y_test, Y_predict)
```

```python
print(classification_report(Y_test, Y_predict))
```
